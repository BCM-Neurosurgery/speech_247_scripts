```python
import numpy as np
import pandas as pd

FPS = 25.0

def moving_average_nan(x: np.ndarray, window: int) -> np.ndarray:
    """
    Moving average that ignores NaNs.
    """
    x = np.asarray(x, dtype=float)
    valid = np.isfinite(x).astype(float)
    x_filled = np.nan_to_num(x, nan=0.0)

    kernel = np.ones(window, dtype=float)
    num = np.convolve(x_filled, kernel, mode="same")
    den = np.convolve(valid, kernel, mode="same")

    out = np.full_like(x, np.nan, dtype=float)
    keep = den > 0
    out[keep] = num[keep] / den[keep]
    return out


def merge_ldasd_tracks(
    tracks,
    scores,
    faces=None,   # kept for compatibility, not required below
    frame_width: int | float = None,
    frame_height: int | float = None,
    fps: float = 25.0,
    smooth_ms: float = 400.0,
    threshold: float = 0.0,
):
    """
    Returns a dataframe with one row per global frame:
      - frame_idx
      - time_sec
      - time_ms
      - raw_score          : max raw score across active tracks
      - smooth_score       : max smoothed score across active tracks
      - has_face           : whether any tracked face is present
      - speaking           : whether any visible face is speaking
      - patient_track      : track id chosen as patient for this frame
      - speaker            : True if patient is speaking, else False

    Patient heuristic:
      among active tracks on each frame, choose the one closest to frame center.
    """
    if len(tracks) != len(scores):
        raise ValueError(f"tracks ({len(tracks)}) and scores ({len(scores)}) length mismatch")

    if frame_width is None or frame_height is None:
        raise ValueError("frame_width and frame_height are required to choose the center-most face")

    # Determine total frame span from all tracks
    max_frame = -1
    for tr in tracks:
        frames = np.asarray(tr["track"]["frame"])
        if len(frames) == 0:
            continue
        max_frame = max(max_frame, int(frames.max()))

    if max_frame < 0:
        raise ValueError("No frames found in tracks")

    n_frames = max_frame + 1
    n_tracks = len(tracks)

    # Per-track matrices over global frames
    raw_score_mat = np.full((n_tracks, n_frames), np.nan, dtype=float)
    x_mat = np.full((n_tracks, n_frames), np.nan, dtype=float)
    y_mat = np.full((n_tracks, n_frames), np.nan, dtype=float)

    # Populate score + center position for each track
    for tidx, (tr, sc) in enumerate(zip(tracks, scores)):
        frames = np.asarray(tr["track"]["frame"], dtype=int)
        sc = np.asarray(sc, dtype=float)
        xs = np.asarray(tr["proc_track"]["x"], dtype=float)
        ys = np.asarray(tr["proc_track"]["y"], dtype=float)

        n = min(len(frames), len(sc), len(xs), len(ys))
        frames = frames[:n]
        sc = sc[:n]
        xs = xs[:n]
        ys = ys[:n]

        raw_score_mat[tidx, frames] = sc
        x_mat[tidx, frames] = xs
        y_mat[tidx, frames] = ys

    # Smooth each track independently
    window = max(1, int(round((smooth_ms / 1000.0) * fps)))
    if window % 2 == 0:
        window += 1

    smooth_score_mat = np.full_like(raw_score_mat, np.nan)
    for tidx in range(n_tracks):
        smooth_score_mat[tidx] = moving_average_nan(raw_score_mat[tidx], window=window)

    # Any face present?
    has_face = np.any(np.isfinite(raw_score_mat), axis=0)

    # Any visible face speaking?
    speaking = np.zeros(n_frames, dtype=int)
    any_active_speaking = np.any(
        np.isfinite(smooth_score_mat) & (smooth_score_mat >= threshold),
        axis=0
    )
    speaking[has_face] = any_active_speaking[has_face].astype(int)

    # Pick patient track = active track closest to frame center
    cx = frame_width / 2.0
    cy = frame_height / 2.0

    dist_mat = np.sqrt((x_mat - cx) ** 2 + (y_mat - cy) ** 2)
    dist_mat[~np.isfinite(raw_score_mat)] = np.nan  # only consider active tracks

    patient_track = np.full(n_frames, -1, dtype=int)
    for f in range(n_frames):
        active = np.where(np.isfinite(dist_mat[:, f]))[0]
        if len(active) == 0:
            continue
        best = active[np.argmin(dist_mat[active, f])]
        patient_track[f] = int(best)

    # Patient speaking?
    patient_raw = np.full(n_frames, np.nan, dtype=float)
    patient_smooth = np.full(n_frames, np.nan, dtype=float)
    speaker = np.zeros(n_frames, dtype=bool)

    for f in range(n_frames):
        tidx = patient_track[f]
        if tidx == -1:
            continue
        patient_raw[f] = raw_score_mat[tidx, f]
        patient_smooth[f] = smooth_score_mat[tidx, f]
        if np.isfinite(patient_smooth[f]) and patient_smooth[f] >= threshold:
            speaker[f] = True

    # Global max score across tracks, for reference
    global_raw = np.nanmax(raw_score_mat, axis=0)
    global_smooth = np.nanmax(smooth_score_mat, axis=0)
    global_raw[~has_face] = np.nan
    global_smooth[~has_face] = np.nan

    df = pd.DataFrame({
        "frame_idx": np.arange(n_frames, dtype=int),
        "time_sec": np.arange(n_frames, dtype=float) / fps,
        "time_ms": (np.arange(n_frames, dtype=float) / fps) * 1000.0,
        "raw_score": global_raw,
        "smooth_score": global_smooth,
        "has_face": has_face.astype(int),
        "speaking": speaking,                 # anyone speaking
        "patient_track": patient_track,       # chosen center-most active track
        "patient_raw_score": patient_raw,
        "patient_smooth_score": patient_smooth,
        "speaker": speaker,                   # True iff patient is speaking
    })

    return df
```

```python
# get speech df for each recording
utterance_dfs = {}

for key, track in tracks.items():
    try:
        score = speaker_scores[key]
        df = merge_ldasd_tracks(
            tracks=track,
            scores=score,
            frame_width=W,
            frame_height=H,
            fps=25.0,
            smooth_ms=400.0,
            threshold=0.5,
        )
        utterance_dfs[key] = df
    except Exception as e:
        print(e)
```

```python
# now get patient speaking bool for each row based on raw score from utterance dfs
from tqdm import tqdm
filter_df["patient_speaking_rough"] = False
groups = filter_df.groupby(['recording_id', 'segment_id', 'source'])
for idx, (info, group) in tqdm(enumerate(groups), total=len(groups)):       
    row = group.iloc[0]
    if row.recording_id not in utterance_dfs:
        continue
    df = utterance_dfs[row.recording_id]
    utterance_mask = (df['time_sec'] >= row.segment_start) & (df['time_sec'] <= row.segment_end)
    # only keep mininum 2 second segments
    seg_len = row.segment_end - row.segment_start
    prop_speaker = (df[utterance_mask]['patient_raw_score'] > 0).sum() / len(df[utterance_mask])
    if prop_speaker >= 0.5 and seg_len >= 2:
        filter_df.loc[group.index.values, 'patient_speaking_rough'] = True
```

```python
# concatenate all rows into audio file
# now get audio source and merge into single audio stream for viewing
sample_rate = 16_000
# get sound array for each word and save to continuous recording path
roommic1_continuous_path = f'{save_path}/RoomMic1_continuous.wav'
# open blank file to write 
# get seg lengths
patient_df = filter_df[(filter_df['patient_speaking_rough'])]
groups = patient_df.groupby(['recording_id', 'segment_id', 'source'])
audio_mats = [None for _ in range(len(groups))]
audio_dict = {}

for idx, (info, group) in enumerate(groups):
    if idx % 10_000 == 0:
        print("processing segment:", idx)
    start_ts = group.iloc[0].segment_start
    end_ts = group.iloc[0].segment_end
    recording_id = group.iloc[0].recording_id
    file_path = f'{data_dir}/{patient}/{recording_id}/audio/RoomMic1_{recording_id}.wav'
    try:
        with sf.SoundFile(file_path, 'r') as f:
            start_frame = int(np.round(start_ts * sample_rate))
            end_frame = int(np.round(end_ts * sample_rate))
            f.seek(start_frame)
            audio_segment_1 = f.read(frames=end_frame - start_frame)
    except Exception as e:
        print(f"Error reading RoomMic1 audio file: {e}")
        audio_segment_1 = np.zeros(end_frame - start_frame)
    audio_mats[idx] = audio_segment_1
    audio_dict[f"{info[0]}_{info[2]}_{info[1]}"] = audio_segment_1
```

```python
# generate continuous between word
continuous_text = np.hstack(audio_mats)
```

```python
# save to evaluate
save_path = f'{data_dir}/{patient}/all_convo_recording'

roommic1_continuous_path = f'{save_path}/RoomMic1_continuous_patient_video.wav'
sf.write(roommic1_continuous_path, continuous_text, sample_rate)
```

```python
# now, index into neural data and see if the segments are equivalent...
# concatenate all rows into audio file
# now get audio source and merge into single audio stream for viewing
from brpylib import NsxFile
sample_rate = 30_000
# get sound array for each word and save to continuous recording path
# open blank file to write 
# get seg lengths
patient_df = filter_df[(filter_df['patient_speaking_rough'])]
groups = patient_df.groupby(['recording_id', 'segment_id', 'source'])
audio_dict = {}
memo = {}

nsp2_audio_chan = 66
nsp1_audio_chan = 2


for idx, (info, group) in enumerate(groups):
    if idx % 10_000 == 0:
        print("processing segment:", idx)
    start_ts = group.iloc[0].segment_start
    end_ts = group.iloc[0].segment_end
    recording_id = group.iloc[0].recording_id
    file_path = f'{data_dir}/{patient}/{recording_id}/neural/{patient}_{recording_id}_NSP-1.ns5'
    try:
        if info[0] in memo:
            audio_sigs = memo[info[0]]
        else:
            io = NsxFile(file_path)
            data = io.getdata()['data']
            audio_sigs = np.hstack([mat[nsp1_audio_chan] for mat in data])
            memo[info[0]] = audio_sigs
        start_sample = int(np.round(start_ts * sample_rate))
        end_sample = int(np.round(end_ts * sample_rate))
        audio_segment_1 = audio_sigs[start_sample:end_sample]
        audio_segment_1 = audio_segment_1 / np.max(np.abs(audio_segment_1))
    except Exception as e:
        print(f"Error reading RoomMic1 audio file: {e}")
        start_sample = int(np.round(start_ts * sample_rate))
        end_sample = int(np.round(end_ts * sample_rate))
        audio_segment_1 = np.zeros(end_sample - start_sample, dtype=np.float64)
    audio_dict[f"{info[0]}_{info[2]}_{info[1]}"] = audio_segment_1
```

```python
# try exporting raw audio
def speaking_segments_from_df(df: pd.DataFrame, min_duration_ms: float = 200.0) -> pd.DataFrame:
    """
    Convert framewise patient-speaking labels into contiguous speaking segments.

    Expects:
      - df["speaker"] to be True/False, where True means the patient is speaking
      - df["time_sec"] to contain absolute time from start of video
    """

    patient_speaking = df["speaker"].to_numpy().astype(int)
    time_sec = df["time_sec"].to_numpy()

    segments = []
    in_seg = False
    start_idx = None

    for i, val in enumerate(patient_speaking):
        if val == 1 and not in_seg:
            in_seg = True
            start_idx = i
        elif val == 0 and in_seg:
            end_idx = i - 1
            segments.append((start_idx, end_idx))
            in_seg = False

    if in_seg:
        segments.append((start_idx, len(patient_speaking) - 1))

    rows = []
    min_duration_sec = min_duration_ms / 1000.0

    for s, e in segments:
        start_sec = time_sec[s]
        end_sec = time_sec[e] + (1.0 / FPS)  # make end exclusive-ish
        dur = end_sec - start_sec

        if dur >= min_duration_sec:
            rows.append({
                "start_frame": s,
                "end_frame": e,
                "start_sec": start_sec,
                "end_sec": end_sec,
                "start_ms": start_sec * 1000.0,
                "end_ms": end_sec * 1000.0,
                "duration_sec": dur,
            })

    return pd.DataFrame(rows)
```

```python
speaker_segs = {}

for key, df in utterance_dfs.items():
    try:
        segs = speaking_segments_from_df(df, min_duration_ms=200.0)
        speaker_segs[key] = segs
    except Exception as e:
        print(e)
        
```

```python
from moviepy import VideoFileClip
stitch_clips = []
for key, seg in speaker_segs.items():
    if seg.empty:
        continue
    print("starting seg", key)
    video_path = video_paths[key]
    video = VideoFileClip(video_path)
    for jdx, row in seg.iterrows():
        clip = video.subclipped(row.start_sec, row.end_sec)
        stitch_clips.append(clip)
```
