```python
# load modules
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import KFold, train_test_split
from sklearn.linear_model import PoissonRegressor
from sklearn.preprocessing import StandardScaler
from scipy.special import gammaln
from scipy.stats import pearsonr, spearmanr
from glob import glob
import os
import json
from collections import deque
```

```python
# set dirs
data_dir = "/mnt/labworlds/Hayden/Hayden_Lab/speech_247/podcast_comparison"
```

```python
# get patient dirs
patient_dirs = glob(f"{data_dir}/Y*")
patient_dirs = {os.path.basename(dir_):dir_ for dir_ in patient_dirs}
```

```python
transcript_dfs = {}
context_size = 50

for pt, dir_ in patient_dirs.items():
    # now compile json into pandas df

    transcript_df = []
    data = None
    with open(f'{dir_}/transcription/podcast_audio/podcast_audio.json', 'r') as f:
        data = json.load(f)
        segment_data = data['segments']
        context_window = deque()
        for idx, segment_info in enumerate(segment_data):
            for word_info in segment_info['words']:
                entry = {}
                if 'start' not in word_info:
                    continue
                entry['word'] = word_info['word']
                entry['start_ts'] = word_info['start']
                entry['end_ts'] = word_info['end']
                entry['qscore'] = word_info['score']
                entry['speaker'] = word_info['speaker'] if 'speaker' in word_info else None
                entry['patient'] = pt
                entry['segment_id'] = idx
                entry['segment_text'] = segment_info['text']
                entry['segment_start'] = segment_info['start']
                entry['segment_end'] = segment_info['end']
                entry['context'] = " ".join(context_window)
                transcript_df.append(entry)
                context_window.append(word_info['word'])
                if len(context_window) > context_size:
                    context_window.popleft()
                
    transcript_df = pd.DataFrame(transcript_df)
    transcript_dfs[pt] = transcript_df
```

```python
# remove jabberwocky
transcript_dfs["YFK"] = transcript_dfs["YFK"].iloc[:6942].copy()
transcript_dfs["YFG"] = transcript_dfs["YFG"].iloc[:6970].copy()
```

```python
for pt, dir_ in patient_dirs.items():
    transcript_df = transcript_dfs[pt]
    transcript_df.to_csv(f'{dir_}/whisperx_out/audio_transcript_df.csv')
```
