```python
from scipy.signal import butter, filtfilt, iirnotch, find_peaks
def bandpass_filter(x, fs, lo=300, hi=6000, order=4, notch_hz=None, q=30):
    ny = 0.5 * fs
    b, a = butter(order, [lo/ny, hi/ny], btype='band')
    xf = filtfilt(b, a, x, axis=-1)
    if notch_hz is not None:
        b_notch, a_notch = iirnotch(notch_hz/ny, q)
        xf = filtfilt(b_notch, a_notch, xf, axis=-1)
    return xf
```

```python
def robust_sigma(x, axis=-1, eps=1e-9):
    med = np.median(x, axis=axis, keepdims=True)
    mad = np.median(np.abs(x - med), axis=axis, keepdims=True) + eps
    return 1.4826 * mad
```

```python
def _detect_peaks_channel(xc, fs, thr, refractory_s, polarity="neg"):
    min_dist = max(1, int(refractory_s * fs))
    if polarity == "neg":
        inv = -xc
        peaks, _ = find_peaks(inv, height=(-thr if thr < 0 else None), distance=min_dist)
    elif polarity == "pos":
        peaks, _ = find_peaks(xc, height=(thr if thr > 0 else None), distance=min_dist)
    elif polarity == "both":
        p_pos, _ = find_peaks(xc, height=(abs(thr) if thr > 0 else None), distance=min_dist)
        inv = -xc
        p_neg, _ = find_peaks(inv, height=(abs(thr) if thr < 0 else None), distance=min_dist)
        spikes = np.sort(np.concatenate([p_pos, p_neg]))
        if spikes.size == 0:
            return spikes
        keep = [spikes[0]]
        for s in spikes[1:]:
            if s - keep[-1] >= min_dist:
                keep.append(s)
        return np.asarray(keep, dtype=int)
    else:
        raise ValueError("polarity must be 'neg', 'pos', or 'both'")
    return peaks.astype(int)
```

```python
def _avg_rate_from_k(xf, fs, k, sigmas, refractory_s, polarity):
    n_ch, n_samp = xf.shape
    total = 0
    for c in range(n_ch):
        thr = (-k * float(sigmas[c, 0])) if polarity == "neg" else (k * float(sigmas[c, 0]))
        peaks = _detect_peaks_channel(xf[c], fs, thr, refractory_s, polarity)
        total += len(peaks)
    duration_s = n_samp / fs
    return total / (duration_s * n_ch)
```

```python
def _solve_k_bisect(xf, fs, target_hz, refractory_s, polarity,
                    k_low=1.5, k_high=10.0, tol_hz=0.5, max_iter=25):
    sigmas = robust_sigma(xf)
    def rate(k):
        return _avg_rate_from_k(xf, fs, k, sigmas, refractory_s, polarity)
    r_low, r_high = rate(k_low), rate(k_high)
    tries = 0
    while r_low < target_hz and tries < 6:
        k_low *= 0.7
        r_low = rate(k_low); tries += 1
    tries = 0
    while r_high > target_hz and tries < 6:
        k_high *= 1.5
        r_high = rate(k_high); tries += 1
    kL, kH = k_low, k_high
    for _ in range(max_iter):
        kM = 0.5 * (kL + kH)
        rM = rate(kM)
        if abs(rM - target_hz) <= tol_hz:
            return kM, sigmas
        if rM > target_hz:
            kL = kM
        else:
            kH = kM
    return 0.5 * (kL + kH), sigmas
```

```python
def find_k_for_target_rate(x, fs, target_hz=20.0, refractory_s=0.001, 
                           k_low=1.5, k_high=10.0, tol_hz=0.5, max_iter=25,
                           use_subset_sec=None, notch_hz=None):
    """
    x: ndarray shape (n_channels, n_samples), raw (unfiltered) voltage in µV
    """
    # Optionally work on a shorter window for speed
    if use_subset_sec is not None:
        n_keep = int(use_subset_sec * fs)
        x = x[:, :min(n_keep, x.shape[1])]
    duration_s = x.shape[1] / fs
    n_ch = x.shape[0]

    # Preprocess and noise scale
    xf = bandpass_filter(x, fs, notch_hz=notch_hz)
    sigmas = robust_sigma(xf)  # shape (n_ch, 1)
    
    # Helper to compute average rate for k
    def avg_rate(k):
        total_spikes = count_spikes_for_k(xf, fs, k, sigmas, refractory_s)
        return total_spikes / (duration_s * n_ch)
    
    # Ensure bounds bracket the solution
    r_low = avg_rate(k_low)
    r_high = avg_rate(k_high)
    # If not bracketed, expand bounds a couple times
    expand = 0
    while r_low < target_hz and expand < 5:
        k_low *= 0.7  # lower threshold -> more spikes
        r_low = avg_rate(k_low); expand += 1
    expand = 0
    while r_high > target_hz and expand < 5:
        k_high *= 1.5  # higher threshold -> fewer spikes
        r_high = avg_rate(k_high); expand += 1

    # Bisection
    for _ in range(max_iter):
        k_mid = 0.5 * (k_low + k_high)
        r_mid = avg_rate(k_mid)
        if abs(r_mid - target_hz) <= tol_hz:
            return k_mid, r_mid
        if r_mid > target_hz:
            # too many spikes -> increase k
            k_low = k_mid
        else:
            # too few spikes -> decrease k
            k_high = k_mid
    # Return best effort
    k_mid = 0.5 * (k_low + k_high)
    return k_mid, avg_rate(k_mid)
```

```python
# ---------- Main API ----------
def detect_and_bin_spikes(
    x_raw,
    fs=30000,
    out_fs=1000,
    refractory_ms=1.0,
    polarity="neg",
    notch_hz=None,
    band_lo=300,
    band_hi=6000,
    k=None,
    target_avg_hz=None,
    tol_hz=0.5,
    k_low=1.5,
    k_high=10.0,
    solve_subset_sec=60
):
    """
    x_raw: ndarray (n_channels, n_samples) raw microvolts.
    Returns:
        spike_bin: (n_channels, n_bins) uint8 binary spike trains at out_fs
        info: dict with thresholds, rates, etc.
    """
    n_ch, n_samp = x_raw.shape
    refractory_s = refractory_ms / 1000.0

    # Filter
    xf_full = bandpass_filter(x_raw, fs, band_lo, band_hi, notch_hz=notch_hz)

    # Solve k if needed
    if k is None:
        if target_avg_hz is None:
            raise ValueError("Provide either k or target_avg_hz.")
        if solve_subset_sec is not None:
            keep = min(int(solve_subset_sec * fs), n_samp)
            xf_solve = xf_full[:, :keep]
        else:
            xf_solve = xf_full
        k, sigmas = _solve_k_bisect(
            xf_solve, fs, target_avg_hz, refractory_s, polarity,
            k_low=k_low, k_high=k_high, tol_hz=tol_hz
        )
    else:
        sigmas = robust_sigma(xf_full)

    # Detect spikes
    spike_indices = []
    for c in range(n_ch):
        thr = (-k * float(sigmas[c, 0])) if polarity == "neg" else (k * float(sigmas[c, 0]))
        peaks = _detect_peaks_channel(xf_full[c], fs, thr, refractory_s, polarity)
        spike_indices.append(peaks)

    # Bin at out_fs
    duration_s = n_samp / fs
    n_bins = int(np.ceil(duration_s * out_fs))
    spike_bin = np.zeros((n_ch, n_bins), dtype=np.uint8)
    for c in range(n_ch):
        if spike_indices[c].size == 0:
            continue
        bins = (spike_indices[c].astype(np.int64) * out_fs) // fs
        bins = bins[bins < n_bins]
        spike_bin[c, np.unique(bins)] = 1

    # Stats
    per_ch_counts = np.array([len(idx) for idx in spike_indices], dtype=float)
    per_ch_rates = per_ch_counts / duration_s
    achieved_avg_hz = per_ch_rates.mean()

    info = {
        "k": float(k),
        "sigmas_uV": sigmas.squeeze(-1),
        "per_channel_rates_Hz": per_ch_rates,
        "achieved_avg_rate_Hz": float(achieved_avg_hz),
        "n_bins": n_bins,
        "out_fs": out_fs,
        "fs": fs,
        "polarity": polarity,
    }
    return spike_bin, info
```

```python
import json
for idx, dir_ in enumerate(recording_dirs):
    print("starting dir_", idx)
    basename = os.path.basename(dir_)
    task_name = basename
    
    # skip folders that already have binary spiketrain
    if os.path.isfile(f'{data_dir}/{patient}/spikes/{task_name}_binary_spiketrain.npy'):
        print("spiketrain already exists for", basename)
        #continue
    try:
        ns5_path = glob(f'{dir_}/*NSP-2.ns5')[0]
    except IndexError as e:
        print(e)
        continue
 
    try:
        # open file and recoridng series
        print("loading ns5")
        rec_file = brpylib.NsxFile(ns5_path)
        rec_data = rec_file.getdata()
        rec_raw = np.hstack(rec_data['data'])
        n_chan, n = rec_raw.shape
        # remove non-sEEG signal
        eeg_idx = np.array(rec_data['elec_ids']) < 500
        rec_raw = rec_raw[eeg_idx]
        out = detect_and_bin_spikes(rec_raw, target_avg_hz=20)

        # save spike trains and concatenate info
        np.save(f'{data_dir}/{patient}/spikes/{task_name}_binary_spiketrain.npy', out[0])
        with open(f'{data_dir}/{patient}/spikes/{task_name}_spike_stats.json', 'w') as fp:
            stats = out[1].copy()
            stats['sigmas_uV'] = stats['sigmas_uV'].tolist()
            stats['per_channel_rates_Hz'] = stats['per_channel_rates_Hz'].tolist()
            json.dump(stats, fp)
    except Exception as e:
        print(e)

    
```
